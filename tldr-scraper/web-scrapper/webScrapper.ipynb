# Web Scrapper for **https://tldr.tech/tech/archives**

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

### Getting All TLDR From Archives

```python
url = "https://tldr.tech/tech/archives"
result = requests.get(url)

soup = bs(result.content)
web_links = soup.select('a') 
actual_web_links = [web_link['href'] for web_link in web_links if not(web_link['href'].find("/tech/")) and len(web_link['href']) == 16]
```

### Getting Content From the First TLDR

```python
url_individual = f"https://tldr.tech{actual_web_links[0]}"
result_individual = requests.get(url_individual)
individual_TLDR = bs(result_individual.content)
```

```python
web_links_individual_TLDR = individual_TLDR.select('a') 
actual_web_links_individual_TLDR = [web_link['href'] for web_link in web_links_individual_TLDR if (str(web_link).find('(') != -1) and (str(web_link).find(')') != -1)]
```

```python
headers_ = individual_TLDR.select('h3')
actual_headers = [str(head)[4:-5].strip() for head in headers_]
```

```python
contents_ = individual_TLDR.select('h3')
actual_content = [str(content.parent.parent.select('div')[0])[5:-6] for content in contents_]
```

```python
MainDataFrame = pd.DataFrame({
    "Links":actual_web_links_individual_TLDR,
    "Headers":actual_headers,
    "Content":actual_content})
```

### Getting IMAGES from Link Websites

```python
dic = {}
for links in actual_web_links_individual_TLDR:
    result = requests.get(links)
    soup = bs(result.content)
    dic[links] = [i["src"] for i in soup.select("img") if i["src"].find("https://") != -1]
```

```python
df = pd.DataFrame(dict([(key, pd.Series(value)) for key, value in dic.items()]))
```

```python
with open('client_secret.json') as user_file:
  file_content = user_file.read()

# convert json to a string
file_content = json.dumps(file_content)

# encode service key
encoded_file_content = base64.b64encode(file_content.encode('utf-8'))

print(encoded_file_content)
# b'many_characters_here'
```

```python
def Create_Service(client_secret_file, api_name, api_version, *scopes):
    CLIENT_SECRET_FILE = client_secret_file
    API_SERVICE_NAME = api_name
    API_VERSION = api_version
    SCOPES = [scope for scope in scopes[0]]
    print(SCOPES)

    cred = None

    pickle_file = f'token_{API_SERVICE_NAME}_{API_VERSION}.pickle'
    # print(pickle_file)

    if os.path.exists(pickle_file):
        with open(pickle_file, 'rb') as token:
            cred = pickle.load(token)

    if not cred or not cred.valid:
        if cred and cred.expired and cred.refresh_token:
            cred.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(CLIENT_SECRET_FILE, SCOPES)
            cred = flow.run_local_server()

        with open(pickle_file, 'wb') as token:
            pickle.dump(cred, token)

    try:
        service = build(API_SERVICE_NAME, API_VERSION, credentials=cred)
        print(API_SERVICE_NAME, 'service created successfully')
        return service
    except Exception as e:
        print('Unable to connect.')
        print(e)
        return None
```

```python
CLIENT_SECRET_FILE = json.loads(base64.b64decode(str(os.getenv("CLIENT_SECRET_FILE"))[2:-1]).decode('utf-8')) #'client_secret.json'
API_SERVICE_NAME = 'sheets'
API_VERSION = 'v4'
SCOPES = ['https://www.googleapis.com/auth/spreadsheets']
gsheetId = '1YXuLzYcUVuB_V_wlCmTPaREcpeKuMK6NYpJ95cRYEr4'

service = Create_Service(CLIENT_SECRET_FILE, API_SERVICE_NAME, API_VERSION, SCOPES)
```
